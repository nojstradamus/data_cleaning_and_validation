### Data cleaning is definitely a "practice makes perfect" skill. Using this dataset of article open-access prices paid by the WELLCOME Trust between 2012 and 2013, determine the five most common journals and the total articles for each. Next, calculate the mean, median, and standard deviation of the open-access cost per article for each journal . 

```python
df = pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv', encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
df = df.drop(columns=['PMID/PMCID', 'Publisher'])
```

```python
df.head()
```

```python
df['Journal title'] = df['Journal title'].apply(lambda x: str(x).lower().capitalize())
```

```python
ja = df.groupby('Journal title').count().drop(columns='COST (£) charged to Wellcome (inc VAT when charged)')
```

```python
ja = ja.sort_values('Article title', ascending=False)
```

```python
ja
```

```python
ja['Article title'].sum()
```

```python
len(df['Article title'].unique())
```

```python
df = df.rename(columns={'COST (£) charged to Wellcome (inc VAT when charged)':'cost'})
df['cost'] = df['cost'].str.replace('£', '')
df['cost'] = df['cost'].str.replace('$', '')
df['cost'] = df['cost'].astype('float64')

print('mean: ', df['cost'].mean())
print('median: ', df['cost'].median())
print('mode: ', df['cost'].mode())
```
